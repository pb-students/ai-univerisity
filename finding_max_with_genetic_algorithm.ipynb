## Znajdowanie maksimum funkcji ) ùëì(ùë•, ùë¶) = ùë•^2 + sin(ùë¶)  w zadanym przedziale przy pomocy algorytmu genetycznego

```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
np.random.seed(0)
```

```python
def f(space):
    return -1 * (space[0]**2+np.sin(space[1])) # domy≈õlnie funkcja geneticalalgoritm znajduje minima,
                                               # dlatego przemna≈ºamy danƒÖ w zadaniu funkcjƒô przez -1, by znale≈∫ƒá maksimum
```

```python
x_bound = np.array([0, 1])
y_bound = np.array([0, np.pi])
varbound = np.array([
    x_bound, 
    y_bound
])
```

```python
default_param = {'max_num_iteration': 20,\
                   'population_size': 20,\
                   'mutation_probability':0.1,\
                   'elit_ratio': 0.1,\
                   'crossover_probability': 0.6,\
                   'parents_portion': 0.3,\
                   'crossover_type':'uniform',\
                   'max_iteration_without_improv': None}
```

```python
def print_ga_param(param_dict):
    print("_"*125)
    print("Parametry modelu: ")
    polish_equivalent = {'max_num_iteration': "Liczba iteracji: ",\
                   'population_size':"Wielko≈õƒá populacji: ",\
                   'mutation_probability': "Prawdopodobie≈Ñstwo mutacji: ",\
                   'elit_ratio': "Liczba osobnik√≥w o najlepszym przystosowaniu przechodzƒÖca do kolejnej populacji: ",\
                   'crossover_probability': "Prawdopodobie≈Ñ≈õtwo krzy≈ºowania: ",\
                   'parents_portion': "Odsetek populacji zape≈Çniony przez poprzedniƒÖ generacjƒô:",\
                   'crossover_type': "Typ krzy≈ºowania: ",\
                   'max_iteration_without_improv': "Mo≈ºliwe iteracje bez poprawy wyniku: "}
    for key in param_dict:
        print(polish_equivalent[key], param_dict[key])
    print("")
```

```python
parameters_dict = {
    'crossover_probability':[0.6, 0.8, 1],
    'mutation_probability': [0.1, 0.2, 0.3, 0.5],
    'population_size':[5, 10, 15, 20],
    'max_num_iteration':[5, 10, 15, 20]
}
```

```python
def train_models(param):
    for val in parameters_dict[param]:
        algorithm_param = default_param.copy()
        algorithm_param[param] = val
        print_ga_param(algorithm_param)
        model=ga(function=f,\
                dimension=2,\
                variable_type='real',\
                variable_boundaries=varbound,\
                algorithm_parameters=algorithm_param)
        model.run()
```

#### Wp≈Çyw prawdopodobie≈Ñstwa krzy≈ºowania na wynik

```python
train_models('crossover_probability')
```

#### Wp≈Çyw prawdopodobie≈Ñstwa mutacji na wynik

```python
train_models('mutation_probability')
```

#### Wp≈Çyw wielko≈õci populacji na wynik

```python
train_models('population_size')
```

#### Wp≈Çyw ilo≈õci iteracji na wynik

```python
train_models('max_num_iteration')
```
